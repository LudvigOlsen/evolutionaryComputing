# Evolutionary audio resembler test bed

```python
import numpy as np
import librosa
import librosa.display
import IPython.display as ipd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_audio(path, sr=16000):
    audio_file, _ = librosa.load(path,
                                 sr=sr,
                                 mono=True)
    return audio_file
```

```python
sr = 16000
```

```python
audio_example_1 = load_audio("/Users/ludvigolsen/Music/Fars musik/03 Going back to America.mp3", sr)[sr:sr*4]
```

```python
ipd.Audio(audio_example_1, rate=sr)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(audio_example_1, sr=sr)
```

```python
initial_slice  = audio_example_1[:2000]
```

```python
ipd.Audio(initial_slice, rate=sr)
```
